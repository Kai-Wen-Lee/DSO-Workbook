# Introduction

### Synthesis vs Analysis
* Synthesis involves decisions, analysis does not
    - How big, what material, what manufacturing method
* Analysis provides information for rational decisions to be made

### Setting up a design search optimisation problem:
* We adopt the ideas of:
    - Objectives
    - Design variables
        - Bounds
    - Constraints
        - Limits
    - Fixed parameters
    - External moise / uncertain parameters
    - Methods of analysis
    - Schemes to link design variables to analysis
    - schemes to link analysis to objectives and constraints

### Key considerations:
* Ability to work with black box codes
* Need for gradient information
* Robustness to poor, noisy or incomplete data or badly shaped functions
* Speed of convergence
* Ability to run calculations in parallel
* Repeatability / average performace of stochastic methods

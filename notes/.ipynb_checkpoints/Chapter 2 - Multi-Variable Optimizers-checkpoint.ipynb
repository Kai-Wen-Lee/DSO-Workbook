### Steepest Descent Approaches

General Idea:
* Simple idea is to look for direction of steepest descent and go in that direction until path stops descending.
* Gradients are re-assessed and process repeats.

I.e.:
1. From a contour plot:


2. Do line search in the direction of steepest descent.
3. Do so until a contour is reached (contour lines mean points of inflection!)
4. Now, the line of steepest descent will be tangent to a contour
5. Due to maths, the next steepest descent will be at 90 degrees to the first S.D.

#### Formally:

$x_{i+1} = x_{i} - \lambda_{i}^* \nabla f(x_{i})$

where
$\nabla f(x_{i})$ are the gradients at $x_{i}$

#### Example:
Minimize $$f_(x_{1},x_{2}) = x_{1} - x_{2} + 2 x_{1}^2 + 2 x_{1} x_{2} + x_{2}^2 $$
Starting at iteration 
$$x_{1} = \begin{bmatrix}0 \\
0
\end{bmatrix}$$


### Problem: 

* Unless our funtion has circular contours, the direction of steepest descent NEVER points at the final optimum
